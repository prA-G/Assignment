# *Theoretical Question*

### 1. What is unsupervised learning in the context of machine learning?

Unsupervised learning is a type of machine learning where the model learns patterns from unlabeled data. The algorithm tries to find structure such as groups, patterns, or associations without any explicit target or label.

### 2. How does K-Means clustering algorithm work?

K-Means works by dividing data into K groups based on similarity. It starts by choosing K initial points called centroids. Each data point is assigned to the nearest centroid. Then the centroids are updated by taking the average of all points in each group. This process of assigning points and updating centroids continues until the centroids stop moving or the model reaches a stable state.

### 3. Explain the concept of a dendrogram in hierarchical clustering.

A dendrogram is a tree-like diagram used in hierarchical clustering. It shows how data points are merged step by step to form clusters. The lower branches show individual points, and the higher branches show how these points join to form larger groups. By cutting the dendrogram at a chosen height, we can select the number of clusters.

### 4.  What is the main difference between K-Means and Hierarchical Clustering?

The main difference is that K-Means needs a fixed number of clusters before starting, while hierarchical clustering builds a tree of clusters and does not require choosing the number of clusters in advance. K-Means uses repeated updates to find cluster centers, but hierarchical clustering forms clusters step by step by merging or splitting groups.

### 5.  What are the advantages of DBSCAN over K-Means?

DBSCAN can find clusters of any shape, while K-Means only works well with round clusters. DBSCAN does not require the number of clusters in advance. It can also identify noise points, which K-Means cannot handle well. It performs better when clusters have different sizes and densities.

### 6. When would you use Silhouette Score in clustering?

Silhouette Score is used to measure how well the data points fit within their assigned clusters. It is used when we want to check the quality of clustering or compare different clustering models.

### 7.  What are the limitations of Hierarchical Clustering?

Hierarchical clustering is slow for large datasets because it has high time complexity. Once a merge or split is done, it cannot be undone. It is also sensitive to noise and outliers. The results can change based on the choice of linkage method and distance measure.

### 8.  Why is feature scaling important in clustering algorithms like K-Means?

Feature scaling is important because K-Means uses distance to form clusters. If features have different scales, the feature with the larger scale will dominate the distance calculation. Scaling makes all features contribute equally, which helps K-Means create better and more balanced clusters.

### 9. How does DBSCAN identify noise points?

DBSCAN marks a point as noise when it does not have enough nearby points within the given distance. If a point is not a core point and also not reachable from any core point, it is treated as noise. Noise points do not belong to any cluster.

### 10.  Define inertia in the context of K-Means.

Inertia is the total distance between each data point and its assigned cluster center. It measures how compact the clusters are. Lower inertia means the points are closer to their centroids, which usually indicates better clustering.

### 11.  What is the elbow method in K-Means clustering?

The elbow method is a technique to choose the right number of clusters in K-Means. It works by plotting the inertia value for different numbers of clusters. The point where the decrease in inertia becomes slow and forms a bend like an elbow is selected as the best number of clusters.

### 12.  Describe the concept of "density" in DBSCAN.

In DBSCAN, density means how many data points lie close to each other in a region. A region is considered dense when a point has enough neighbors within a fixed distance. Dense areas form clusters, while areas with low density are treated as noise.

### 13.  Can hierarchical clustering be used on categorical data?

Yes, hierarchical clustering can be used on categorical data, but it requires a suitable distance measure. Standard distance measures like Euclidean do not work well, so methods such as Hamming distance or specialized algorithms must be used. Without proper distance handling, the results may not be reliable.

### 14.  What does a negative Silhouette Score indicate?

A negative Silhouette Score indicates that many data points are placed in the wrong clusters. It means a point is closer to a different cluster than the one it is assigned to. This shows poor clustering quality and incorrect group formation.

### 15.  Explain the term "linkage criteria" in hierarchical clustering.

Linkage criteria define how the distance between two clusters is calculated in hierarchical clustering. It decides how clusters are merged at each step. Common linkage types include single linkage, complete linkage, and average linkage. Each method uses a different way to measure the distance between clusters.

### 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

K-Means performs poorly when clusters have different sizes or densities because it assumes all clusters are similar in shape and spread. It uses distance to assign points, so larger or denser clusters can pull centroids toward them. This causes incorrect grouping and inaccurate cluster boundaries.

### 17.  What are the core parameters in DBSCAN, and how do they influence clustering?

DBSCAN has two main parameters: eps and min_samples.
eps sets the maximum distance to consider two points as neighbors.
min_samples decides how many neighbors a point must have to be a core point.
A larger eps forms bigger clusters, while a smaller eps forms tighter clusters.
A higher min_samples makes it harder for a point to become a core point, which may increase noise points.

### 18.  How does K-Means++ improve upon standard K-Means initialization?

K-Means++ improves K-Means by choosing better initial centroids. Instead of selecting all centroids randomly, it picks the first centroid randomly and then selects the next centroids based on distance.

### 19. What is agglomerative clustering?

Agglomerative clustering is a type of hierarchical clustering that builds clusters from the bottom up. It starts with each data point as its own cluster and then repeatedly merges the closest clusters. This process continues until all points form a single large cluster or until the desired number of clusters is reached.

### 20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Silhouette Score is better than inertia because it considers both compactness of clusters and separation between clusters. Inertia only measures how close points are to their centroids and does not check if clusters overlap

### 27.  Generate synthetic data using make_blobs with varying cluster standard deviations and cluster with DBSCAN.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# create blobs: three centers with different spreads
X, y_true = make_blobs(n_samples=500,
                       centers=[[0, 0], [5, 5], [0, 6]],
                       cluster_std=[0.2, 1.6, 0.6],
                       random_state=42)

# scale features (DBSCAN is sensitive to scale)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# apply DBSCAN
db = DBSCAN(eps=0.6, min_samples=6)
labels = db.fit_predict(X_scaled)

# report cluster counts (exclude noise label -1)
unique_labels = np.unique(labels)
n_clusters = len(unique_labels[unique_labels != -1])
n_noise = np.sum(labels == -1)
print("Estimated clusters (excluding noise):", n_clusters)
print("Noise points:", int(n_noise))
print("Cluster sizes:")
for lab in unique_labels:
    if lab == -1:
        continue
    print(f" Cluster {lab}: {int(np.sum(labels == lab))}")

# plot clusters and noise
plt.figure(figsize=(7, 5))
core_mask = labels != -1
plt.scatter(X[core_mask, 0], X[core_mask, 1], c=labels[core_mask], s=30)
if np.any(labels == -1):
    plt.scatter(X[labels == -1, 0], X[labels == -1, 1], c='red', s=20, label='noise')
plt.title("DBSCAN on make_blobs with varying cluster std")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(loc='best')
plt.show()
```

### 29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# generate synthetic data
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.8, random_state=42)

# evaluate silhouette scores for k = 2 to 5
ks = range(2, 6)
scores = []
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)
    print(f"k = {k}, Silhouette Score = {score:.4f}")

# plot bar chart
plt.figure(figsize=(6,4))
plt.bar([str(k) for k in ks], scores)
plt.title("Silhouette Score for k = 2 to 5")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
for i, v in enumerate(scores):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center')
plt.ylim(0, max(scores) + 0.1)
plt.show()
```

### 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# generate overlapping clusters
X, y = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)

# apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# create meshgrid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 400),
                     np.linspace(y_min, y_max, 400))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# plot decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=20)
plt.scatter(centers[:, 0], centers[:, 1], color='black', s=80)
plt.title("K-Means Decision Boundaries on Overlapping Blobs")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

### 34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot


```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# load dataset
data = load_breast_cancer()
X = data.data

# scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# calculate inertia for k = 2 to 6
ks = range(2, 7)
inertias = []

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    print(f"k = {k}, Inertia = {kmeans.inertia_:.2f}")

# plot inertia values
plt.plot(list(ks), inertias, marker='o')
plt.title("Inertia Values for K = 2 to 6 (Breast Cancer Dataset)")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

### 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd            
import seaborn as sns          
import matplotlib.pyplot as plt

# load wine dataset and select first 4 features
data = load_wine()
X = data.data[:, :4]
feature_names = data.feature_names[:4]

# apply KMeans (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

# prepare DataFrame for pairplot
df = pd.DataFrame(X, columns=feature_names)
df['cluster'] = labels.astype(str)  # use string so seaborn shows discrete hue

# plot pairplot of first 4 features colored by cluster
sns.pairplot(df, vars=feature_names, hue='cluster', diag_kind='kde', corner=False)
plt.suptitle("Wine dataset - KMeans cluster assignments (first 4 features)", y=1.02)
plt.show()
```

### 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# generate noisy blobs
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=1.2, random_state=42)

# apply DBSCAN
db = DBSCAN(eps=0.9, min_samples=5)
labels = db.fit_predict(X)

# count clusters (excluding noise)
unique_labels = np.unique(labels)
clusters = [lab for lab in unique_labels if lab != -1]
num_clusters = len(clusters)

# count noise points
noise_count = np.sum(labels == -1)

print("Number of clusters (excluding noise):", num_clusters)
print("Number of noise points:", noise_count)
```

### 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np

# load digits dataset
data = load_digits()
X = data.data

# reduce to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_2d = tsne.fit_transform(X)

# apply Agglomerative Clustering (use 10 clusters for 10 digits)
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_2d)

# plot clusters
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, s=15, cmap='tab10')
plt.title("Agglomerative Clustering on Digits after t-SNE")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.colorbar(scatter, ticks=range(10), label='cluster label')
plt.tight_layout()
plt.show()

# print cluster sizes
unique, counts = np.unique(labels, return_counts=True)
for u, c in zip(unique, counts):
    print(f"Cluster {u} size: {c}")
```
